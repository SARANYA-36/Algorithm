```python
import time
import matplotlib.pyplot as plt
def linear_search(arr, target):
    for i, num in enumerate(arr):
        if num == target:
            return i
    return -1
def measure_time(arr, target):
    start_time = time.time()
    linear_search(arr, target)
    end_time = time.time()
    return end_time - start_time
def experiment(n_values, target):
    time_taken = []
    for n in n_values:
        arr = list(range(n))
        time_taken.append(measure_time(arr, target))
    return n_values, time_taken
if __name__ == "__main__":
    n_values = [10, 100, 500, 1000, 2000, 5000, 10000]
    target = 500
    n_values, time_taken = experiment(n_values, target)
    plt.figure(figsize=(8, 5))
    plt.plot(n_values, time_taken, marker='o', linestyle='-', color='r', label="Linear Search")
    plt.xlabel("Number of Elements (n)")
    plt.ylabel("Time Taken (seconds)")
    plt.title("Linear Search Time Complexity")
    plt.grid(True)
    plt.legend()
    plt.show()
```
