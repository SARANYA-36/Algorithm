```python
import time
import matplotlib.pyplot as plt
def recursive_binary_search(arr,low,high,x):
  if high >= low:
    mid = (high + low ) // 2
    if arr[mid] == x:
      return mid
    elif arr[mid] > x:
      return recursive_binary_search(arr,low,mid -1,x)
    else:
      return recursive_binary_search(arr,mid+1,high,x)
  else:
    return -1
def measure_search_time(arr,x):
  start_time = time.time()
  result = recursive_binary_search(arr,0,len(arr) -1,x)
  end_time = time.time()
  return end_time - start_time,result
def run_experiments(n_values,x):
  times =[]
  for n in n_values:
    arr = sorted(list(range(n)))
    time_taken,_ = measure_search_time(arr,x)
    times.append(time_taken)
  plt.plot(n_values,times)
  plt.xlabel("Number of Elements(n)")
  plt.ylabel("TimeTaken(seconds)")
  plt.title("Time complexity of recursive Binary Search")
  plt.show()
if __name__ == "__main__":
  n_values = [10,100,1000,10000,100000]
  x = 50
  run_experiments(n_values,x)
```
