```python
import time
import random
import heapq
import matplotlib.pyplot as plt
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
def heap_sort(arr):
    heapq.heapify(arr)
    sorted_arr = [heapq.heappop(arr) for _ in range(len(arr))]
    arr[:] = sorted_arr
def measure_time(sort_function, arr):
    start_time = time.time()
    sort_function(arr)
    end_time = time.time()
    return end_time - start_time
sizes = [100, 500, 1000, 5000, 10000, 20000, 50000]
insertion_times = []
heap_times = []
for n in sizes:
    random_list = [random.randint(0, 100000) for _ in range(n)]
    insertion_list = random_list.copy()
    insertion_times.append(measure_time(insertion_sort, insertion_list))
    heap_list = random_list.copy()
    heap_times.append(measure_time(heap_sort, heap_list))
plt.figure(figsize=(10, 5))
plt.plot(sizes, insertion_times, marker='o', linestyle='-', label='Insertion Sort', color='red')
plt.plot(sizes, heap_times, marker='s', linestyle='-', label='Heap Sort', color='blue')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Sorting Time Complexity: Insertion Sort vs Heap Sort')
plt.legend()
plt.grid()
plt.show()
```
